```python
import numpy as np 
from numpy import exp
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
```

```python
matplotlib.style.available
```

```python
#sns.set_theme(style='ticks',palette='bright')
matplotlib.style.use('classic')
```

```python
def plot_N(N0,lamda,t):
    return N0*exp(-lamda*t)
```

```python
N0=1
lambda_1=1
lambda_2=0.8
lambda_3=0.5
time = np.arange(0,6,0.1)
```

```python
Nt_1=[]
Nt_2=[]
Nt_3=[]
for t in time:
    Nt_1.append(plot_N(N0,lambda_1,t))
    Nt_2.append(plot_N(N0,lambda_2,t))
    Nt_3.append(plot_N(N0,lambda_3,t))
```

```python
plt.plot(time,Nt_1,time,Nt_2,time,Nt_3)
plt.title('Grafik Peluruhan')
plt.legend(['$\lambda = 1$', '$\lambda = 0.8$', '$\lambda = 0.5$'])
plt.ylabel('$N_t$',fontsize=16)
plt.xlabel('time')
```
